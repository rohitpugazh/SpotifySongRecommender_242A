```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Get data and inspect it

data = pd.read_csv('tracks.csv')
data.head()
```

```python
# Percentage of missing values
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100
print("\nPercentage of Missing Values:\n", missing_percentage)
```

```python
client_id = '7c6080dc8b7d4db7a3420649a1cad63a'
client_secret = '616f4520625749488f9e2369f1f12402'

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
def get_artist_name(track_id):
    try:
        track_info = sp.track(track_id)
        artists = track_info['artists']
        artist_names = [artist['name'] for artist in artists]
        return artist_names
    except spotipy.exceptions.SpotifyException as e:
        print(f"Error fetching artist names for track {track_id}: {e}")
        return None  # Return empty on error
```

```python
# Example usage:
track_id = '5Zg3duahoMiGRHScIXVTJu'
artist_names = get_artist_name(track_id)
print(artist_names)
```

```python
for index, row in data.iterrows():
    track_id = row['track_id']
    track_artists = row['track_artists']

    # Check if track_artists is empty
    if not track_artists:
        print(f"Fetching artist names for track {track_id}")
        artist_names = get_artist_name(track_id)
        if artist_names:
            print(f"Found artist names: {artist_names} for track {track_id}")
            data.at[index, 'track_artists'] = artist_names
        else:
            print(f"Failed to find artist names for track {track_id}")
    # If track_artists is a valid list, skip it
    elif isinstance(track_artists, list):
        continue
```

```python
data.to_csv('updated_tracks.csv')
```
